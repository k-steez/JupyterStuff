```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return x**2 + 1

# Function for Left Riemann Sum
def left_riemann(n):
    x = np.linspace(0, 2, n+1)  # n+1 points for n subintervals
    x_left = x[:-1]  # Left endpoints
    dx = 2 / n
    left_sum = sum(f(x_left) * dx)  # Compute sum correctly
    return left_sum

# Define parameters
n = 10
dx = 2 / n
x = np.linspace(0, 2, n+1)  # n+1 points
x_left = x[:-1]  # Left endpoints
f_left = f(x_left)

# Compute left Riemann sum
approx_integral = left_riemann(n)
true_integral = 14/3  # Actual integral result

# Plot function
x_curve = np.linspace(0, 2, 100)  # Smooth curve
plt.plot(x_curve, f(x_curve), color='magenta', label='f(x) = x^2 + 1')

# Plot left Riemann sum
plt.bar(x_left, f_left, width=dx, alpha=0.2, align='edge', edgecolor='blue', label='Left Riemann Sum')
plt.scatter(x_left, f_left, color='c', marker='.', s=100, label='Left Points')

# Compute and plot absolute error
y_diff = np.abs(f_left - (true_integral - approx_integral))
plt.plot(x_left, y_diff, label='Absolute Error', color='orange', marker='*')

plt.legend()
plt.show()
```
