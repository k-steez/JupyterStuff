Using Numpy:
1. Create the discrete interval $x \in [-2\pi,2\pi]$ with 301 elements.
2. Create the function $y_1 = \cos(2x)$
3. create the function $y_2 = \cos(4x)$
4. Create the function $y_3 = e^{-x}\cos(6x)$
5. Plot all three functions in the same diagram
6. You learnt in calculus the differentiation formula $$f^\prime = \lim_{h\rightarrow 0}\frac{f(x+h)-f(x)}{h}. $$ We can use a discrete approximation $$f^\prime = \frac{f(x+h)-f(x)}{h}. $$ Using this formula, approimate the derivative of $y_2$ using $h = 0.1, 0.01, 0.001.$ Plot the approximations and the actual value of the derivative on the same plot. What happens to the approximation as $h$ gets smaller?

```python
import numpy as np
```

```python
x = np.linspace(-2*np.pi,2*np.pi,301)
```

```python
y1 = np.cos(2*x)
y2 = np.cos(4*x)
y3 = np.exp(-x)*np.cos(6*x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y1,label='cos(2x)')
plt.plot(x,y2,label='cos(4x)')
plt.legend(loc='lower right')
plt.figure()
plt.plot(x,y3)
```

```python
def f(x):
    return np.cos(4*x)

def f_derivative(x,h):
    return (f(x+h)-f(x))/h

h_values = [0.1,0.01,0.001]
x_values = np.linspace(-2*np.pi,2*np.pi,301 )
y_values = [f_derivative(x_values, h) for h in h_values]
dy2 = -4*np.sin(4*x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x_values,y_values[0], label = 'h= 0.1', color ='red')
plt.plot(x_values,y_values[1], label = 'h= 0.01', color = 'green')
plt.plot(x_values,y_values[2], label = 'h= 0.001', color = 'black')
plt.title('approximate derivative of y=cos(4x)')
plt.plot(x_values,dy2, label = 'actual derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,301)
h = [0.1,0.01,0.001]
y_vals = []
for i in h:
    y = f_derivative(x,i)
    y_vals.append(y)
    plt.plot(x,y,label=str(i))
plt.plot(x,dy2,label='Actual')
plt.legend(loc = 'lower right')
plt.title('approximate derivative of y=cos(4x)')
#plt.plot([1.25,1.25],[0,4])
#plt.xlim(0,2)
#plt.ylim(3,4)

plt.figure()
y_diffs = []
for i in range(3):
    y_diff = np.absolute(y_vals[i]-dy2)
    y_diffs.append(y_diff)
    plt.plot(x,y_diff,label = str(h[i]))
plt.legend()

plt.figure()
e_max = [i.max() for i in y_diffs]
plt.plot(h,e_max)
plt.plot(h,e_max,marker='*')
```
