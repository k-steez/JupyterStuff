```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 + 1
```

```python
def left_riemann(n):
    x = np.linspace(0,2,n+1)   
    x_left = x[:-1]
    diff = 2/n
    for i in range(n):
        left_sum = sum(f(x_left) * diff)
    return left_sum

def interval(n):
    return 2/n

n = 10
approx_integral = left_riemann(n)
actual_integral = 14/3

x = np.linspace(0, 2, n+1)  
x_left = x[:-1]  
f_left = f(x_left)






```

```python
plt.plot(x,f(x), color = 'magenta', label = 'x^2 + 1')
plt.plot(x_left, f_left, 'c.', markersize = 10, label = 'Left Riemann points')
plt.bar(x_left, f_left, width= interval(10) , alpha= 0.5 , align = 'edge', edgecolor = 'black')


y_diff = np.absolute(f_left-(actual_integral - approx_integral))
y_diff2 = f_left-(actual_integral - approx_integral)

plt.plot(x_left,y_diff, label ='Absolute Difference', color = 'orange', marker = '*')



plt.legend()
plt.show()



```
