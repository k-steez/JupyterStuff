Using Numpy:
1. Create the discrete interval $x in [-2/pi,2/pi]$ with 301 elements
2. Create the function $y_1 = \cos(2x)$
3. Create the function $y_1 = \cos(4x)$
4. Create the function $y_3 = e^(-x)\cos(6x)$
5. Plot all three functions in the same diagram.
6. You learnt in calculus the differentiation formula $$f^\prime = \lim_{h\rightarrow 0}\frac{f(x+h) - f(x)}{h}.$$ We can use a discrete approximation $$f^\prime = \frac{f(x+h) - f(x)}{h}$$. Using this formula, approximate the derivative of $y_2$ using $h = 0.1,0.01, 0.001$ Plot the approximations and the actual value of the derivative on the same plot. What happens to the approximation as $h$ gets smaller?

```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 301)
```

```python
y1 = np.cos(2*x)
y2 = np.cos(4*x)
y3 = np.e**(-x)*np.cos(6*x)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.figure()
plt.plot(x,y3)
```

```python



dev_1 = (np.cos(4*(x + 0.1)) - np.cos(4*x))/0.1
dev_2 = (np.cos(4*(x + 0.01)) - np.cos(4*x))/0.01
dev_3 = (np.cos(4*(x + 0.001)) - np.cos(4*x))/0.001
dev_actual = -4*np.sin(4*x) 
```

```python
import numpy as np
import math as mt
import matplotlib.pyplot as plt

plt.plot(x,dev_1, color= 'green')
plt.plot(x,dev_2, color = 'blue')
plt.plot(x,dev_3, color = 'red')
plt.plot(x,dev_actual)

##plt.plot(x,dev_actual, marker = '*', markersize = 10)
## plt.xlim(0,2)
###plt.ylim(3,4)
## plt.plot([1,0],[0,4])


##To do: Revise markers(edge color,face color, linestyles, linewidth, web face colors and matplotlib colors). read matplotlib docs
```
