Create an array to represent the function $$ y = e^{-\left(x/2\right)^2} $$ for $x\in [-2,2]$ with 151 points in the array. The exponential function is represented as the function `exp` in numpy

Also plot $x$ against $y$ using dashed lines. Use any color of your choice. Your plot should be given the title "Plot of exponential function." The function plt.title can be used for it. You should also add an $x-$axis label "x-axis" using the function plt.xlabel and $y-$axis label "y-axis" using the function plt.ylabel

```python
x = np.linspace(-2,2,151)
y = np.exp(-(x/2)**2)
plt.plot(x,y,linestyle = '--', color = "indigo")
plt.title('Plot of exponential function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
           
```

Using the functions $f(x)=x^2-x-6$ and $g(x)=e^{x/2},$ on the interval $x\in[-3,4]$ with 201 elements, show that the problem of finding $f(x)=g(x)$ is the same as finding $f(x)-g(x)=0$ and $g(x)-f(x)=0.$ Note: You can plot all three functions and use vertical lines to show the relationship between the intersections and roots.

```python
x = np.linspace(-3,4,201)
f = x**2 - x - 6
g = np.exp(x/2)
plt.plot(x,f, label = 'f')
plt.plot(x,g ,label = 'g')
plt.plot(x,f-g, label = 'f(x)- g(x)')
plt.plot(x,g-f, label = 'g(x) - f(x)')
plt.plot([2.1967224577,2.1967224577], [-8,8], label = 'root', linestyle = '--')
plt.plot([-2.05,-2.05],[-8,8], label = 'line of intersection', linestyle = '--')
plt.legend(loc = 'lower right')

```

Plot the parametric curve defined by $$x=16\sin^3(t),$$ $$y=13\cos(t)-5\cos(2t)-2\cos(3t)-\cos(4t)$$ for $t\in[0,2\pi].$
